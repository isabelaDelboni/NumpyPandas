```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/isabe/Documents/Numpy/Salary_Data.csv")
```

```python
data = df[['YearsExperience', 'Salary']].to_numpy()

#Tirar do formato cientifico
np.set_printoptions(suppress=True, precision = 1)


print("Array: \n", data)
```

```python

mediaSalario = np.mean(data[: , 1])  

print("Média de salários: ", mediaSalario)
```

```python
mediaServico = np.mean(data[: , 0])  

print("Média de tempo de serviço: ", mediaServico)
```

```python
funcionariosMaior5 = data[data[:, 0] > 5]
mediaSalarioMaior5 = np.mean(funcionariosMaior5[:, 1])
mediaSalarioMaior5 = round(mediaSalarioMaior5, 2)

print("Média de salários dos funcionários com mais de 5 anos de experiência: ", mediaSalarioMaior5)
```

```python
anosExperiencia = data[:, 0]
salarios = data[:, 1]


plt.figure(figsize=(12, 6))
posicoes = np.arange(len(anosExperiencia)) 

plt.bar(posicoes, salarios, color='yellow', edgecolor='black', alpha=0.7, width=0.8)

plt.xlabel('Anos de Experiência', fontsize=14)
plt.ylabel('Salário', fontsize=14)

plt.tight_layout() 
plt.show()
```

```python
anosExperiencia = data[:, 0]
salarios = data[:, 1]

novosSalarios = salarios * 1.10

resultadoAumento = pd.DataFrame({
    'Anos de Experiência': anosExperiencia,
    'Salário Atual': salarios,
    'Novo Salário': novosSalarios
})

print(resultadoAumento)


plt.figure(figsize=(12, 6))
posicoes = np.arange(len(anosExperiencia))

plt.bar(posicoes - 0.2, salarios, width=0.4, label='Salário Atual', color='blue', edgecolor='black')
plt.bar(posicoes + 0.2, novosSalarios, width=0.4, label='Novo Salário (10% a mais)', color='gray', edgecolor='black')

plt.xlabel('Anos de Experiência', fontsize=14)
plt.ylabel('Salário', fontsize=14)
plt.legend()

plt.tight_layout() 
plt.show()
```
